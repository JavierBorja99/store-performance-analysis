### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()


print(f"Facturación total Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Facturación total Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Facturación total Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Facturación total Tienda 4: ${facturacion_tienda4:,.2f}")

```

```python
import matplotlib.pyplot as plt
```

```python
facturacion = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}

plt.figure(figsize=(8, 5))
plt.bar(facturacion.keys(), facturacion.values(), color='skyblue')
plt.title('Facturación Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Facturación ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python

categorias_t1 = tienda['Categoría del Producto'].value_counts()
categorias_t2 = tienda2['Categoría del Producto'].value_counts()
categorias_t3 = tienda3['Categoría del Producto'].value_counts()
categorias_t4 = tienda4['Categoría del Producto'].value_counts()


print("Tienda 1 - Categorías más populares:\n", categorias_t1, "\n")
print("Tienda 2 - Categorías más populares:\n", categorias_t2, "\n")
print("Tienda 3 - Categorías más populares:\n", categorias_t3, "\n")
print("Tienda 4 - Categorías más populares:\n", categorias_t4)
```

```python
import matplotlib.pyplot as plt

categorias_t1.plot(kind='bar', figsize=(10, 5), color='green')
plt.title("Tienda 1 - Productos vendidos por categoría")
plt.xlabel("Categoría del Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()
```

```python
print(f"Promedio calificación tienda 1: {promedio_t1:.2f}")
print(f"Promedio calificación tienda 2: {promedio_t2:.2f}")
print(f"Promedio calificación tienda 3: {promedio_t3:.2f}")
print(f"Promedio calificación tienda 4: {promedio_t4:.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]

plt.figure(figsize=(8, 5))  # Tamaño del gráfico
plt.bar(nombres_tiendas, promedios, color='orange')
plt.title('Promedio de Calificación por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)  # Porque la calificación suele ir de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()
```

```python
envios = {
    "Tienda 1": envio_t1,
    "Tienda 2": envio_t2,
    "Tienda 3": envio_t3,
    "Tienda 4": envio_t4
}

envios_df = pd.DataFrame.from_dict(envios, orient='index', columns=['Costo Promedio de Envío'])
print(envios_df)
```

```python
envios_df.plot(kind='bar', legend=False, color='skyblue')
plt.title("Costo promedio de envío por tienda")
plt.ylabel("Costo promedio")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
